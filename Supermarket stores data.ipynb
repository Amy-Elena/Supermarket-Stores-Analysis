**Plots**

**Comparing *items available* and *store sales***

```python
df.plot.scatter(x="Items_Available", y="Store_Sales")
plt.xlabel('Items Available')
plt.ylabel('Store Sales')
plt.title('Items Available vs Store Sales')
plt.show()
```

**Comparing *store area* and *items available***

```python
df.plot.scatter(x="Store_Area", y="Items_Available")
plt.xlabel('Store_Area')
plt.ylabel('Items Available')
plt.title('Store Area vs Items Available')
plt.show()
```

**Comparing *daily customer count* and *store sales***

```python
df.plot.scatter(x="Daily_Customer_Count", y="Store_Sales")
plt.xlabel('Daily Customer Count')
plt.ylabel('Store Sales')
plt.title('Daily Customers vs Store Sales')
plt.show()
```

**Comparing *area* and *store sales***

```python
df.plot.scatter(x="Store_Area", y="Store_Sales")
plt.xlabel('Store_Area')
plt.ylabel('Store Sales')
plt.title('Store_Area vs Store Sales')
plt.show()
```

**Comparing *store area* and *customer count***

```python
df.plot.scatter(x="Store_Area", y="Daily_Customer_Count")
plt.xlabel('Store_Area')
plt.ylabel('Daily_Customer_Count')
plt.title('Store_Area vs Customer Count')
plt.show()
```

**Plotting all variables**
